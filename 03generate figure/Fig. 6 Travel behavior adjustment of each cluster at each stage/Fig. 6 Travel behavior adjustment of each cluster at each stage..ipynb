```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import MultipleLocator
import matplotlib as mp
import warnings
warnings.filterwarnings("ignore")
from matplotlib.ticker import FormatStrFormatter, StrMethodFormatter
import locale
locale.setlocale(locale.LC_ALL, "deu_deu")
import matplotlib as mpl
mpl.rcParams['axes.formatter.use_locale'] = True
plt.rc('font',family='Times New Roman') 
import matplotlib.ticker as tkr
def func1(x, pos):  # formatter function takes tick label and tick position
    s = '{:0,}'.format(x)
    return s

def func2(x, pos):  # formatter function takes tick label and tick position
    s = '{:0,}'.format(int(x))
    return s

y_format1 = tkr.FuncFormatter(func1)  # make formatter
y_format2 = tkr.FuncFormatter(func2)  # make formatter
```

# 1.First and last detected time distribution

```python
AllTab = pd.read_csv("first_last_dec_cluster0.csv" , encoding='gbk')
fig, (ax1, ax2 , ax3) = plt.subplots(3, 1, sharex=True, figsize=(2.4, 4),dpi=600,constrained_layout=True)
plt.subplots_adjust(wspace =0, hspace =0.06)
# plt.gca()
font = {'family' : 'Times New Roman',
       'weight' : 'light',
       'size'   : '6'}
# plt.rc('font', **font)  # pass in the font dict as kwargs
plt.rc('axes',unicode_minus=False)
x_major_locator=MultipleLocator(2)
ax1.xaxis.set_major_locator(x_major_locator)
ax1.set_ylim(0,1.1)
sns.kdeplot(AllTab['ft1'],cumulative=True , ax = ax1 , color='green',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['ft2'],cumulative=True , ax = ax1 , color = 'red',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['ft4'],cumulative=True , ax = ax1 , color = 'black',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt1'],cumulative=True , ax = ax1,  linestyle='dashed',color='green',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt2'],cumulative=True , ax = ax1 ,  linestyle='dashed',color = 'red',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt4'],cumulative=True , ax = ax1,  linestyle='dashed',color = 'black',bw=0.18, linewidth=0.5)
ax1.set_ylabel("Cumulative probability",font)
# plt.xlim(0,24)
ax1.legend(['First detected time in the first stage','First detected time in the second stage',
            'First detected time in the third stage','Last detected time in the first stage',
            'Last detected time in the second stage','Last detected time in the third stage'],
          fontsize=6,frameon=False,bbox_to_anchor=(0, 1.62), loc=2, borderaxespad=0,ncol=1,columnspacing=0.22,labelspacing=0.2)
ax1.tick_params(labelsize=6,width=0.3,size=1.5)
bwith=0.3
ax1.spines['bottom'].set_linewidth(bwith)
ax1.spines['left'].set_linewidth(bwith)
ax1.spines['top'].set_linewidth(bwith)
ax1.spines['right'].set_linewidth(bwith)
ax1.yaxis.set_major_formatter(y_format1)

AllTab = pd.read_csv("first_last_dec_cluster1.csv" , encoding='gbk')
sns.kdeplot(AllTab['ft1'],cumulative=True , ax = ax2 , color='green',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['ft2'],cumulative=True , ax = ax2 , color = 'red',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['ft4'],cumulative=True , ax = ax2 , color = 'black',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt1'],cumulative=True , ax = ax2,  linestyle='dashed',color='green',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt2'],cumulative=True , ax = ax2 ,  linestyle='dashed',color = 'red',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt4'],cumulative=True , ax = ax2,  linestyle='dashed',color = 'black',bw=0.18, linewidth=0.5)
# ax2.set_xlabel("First and last detected time of each stage")
ax2.set_ylabel("Cumulative probability",font)
ax2.set_ylim(0,1.1)
# ax2.legend(['First detected time in the first stage','First detected time in the second stage',
#             'First detected time in the third stage','Last detected time in the first stage',
#             'Last detected time in the second stage','Last detected time in the third stage'],
#           loc = 'upper left' ,fontsize=4.2,frameon=False)
ax2.tick_params(labelsize=6,width=0.3,size=1.5)
ax2.spines['bottom'].set_linewidth(bwith)
ax2.spines['left'].set_linewidth(bwith)
ax2.spines['top'].set_linewidth(bwith)
ax2.spines['right'].set_linewidth(bwith)
ax2.yaxis.set_major_formatter(y_format1)

AllTab = pd.read_csv("first_last_dec_cluster2.csv" , encoding='gbk')
sns.kdeplot(AllTab['ft1'],cumulative=True , ax = ax3 , color='green',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['ft2'],cumulative=True , ax = ax3 , color = 'red',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['ft4'],cumulative=True , ax = ax3 , color = 'black',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt1'],cumulative=True , ax = ax3,  linestyle='dashed',color='green',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt2'],cumulative=True , ax = ax3 ,  linestyle='dashed',color = 'red',bw=0.18, linewidth=0.5)
sns.kdeplot(AllTab['lt4'],cumulative=True , ax = ax3,  linestyle='dashed',color = 'black',bw=0.18, linewidth=0.5)
ax3.set_xlabel("First and last detected time of each stage",font)
ax3.set_ylabel("Cumulative probability",font)
ax3.set_ylim(0,1.1)
plt.xlim(0,24)
# ax3.legend(['First detected time in the first stage','First detected time in the second stage',
#             'First detected time in the third stage','Last detected time in the first stage',
#             'Last detected time in the second stage','Last detected time in the third stage'],
#           loc = 'upper left' ,fontsize=4.2,frameon=False)
ax3.tick_params(labelsize=6,width=0.3,size=1.5)
ax3.spines['bottom'].set_linewidth(bwith)
ax3.spines['left'].set_linewidth(bwith)
ax3.spines['top'].set_linewidth(bwith)
ax3.spines['right'].set_linewidth(bwith)
ax3.yaxis.set_major_formatter(y_format1)
plt.show()
```
